```python
import random
import matplotlib.pyplot as plt
```

```python

def simular_galton(num_canicas, num_niveles):
    # Iniciamos una lista de contenedores desde ceros
    contenedores = [0] * (num_niveles + 1)

    # Realizamos la simulación para cada canica
    for _ in range(num_canicas):
        # Agregar la posición de la canica en el nivel 0
        posicion = 0
         # Simulamos la caída de la canica a través de los niveles
        for _ in range(num_niveles):
            # Elegimos aleatoriamente si la canica cae a la izquierda (-1) o a la derecha (1)
            direccion = random.choice([-1, 1])
            posicion += direccion
        # Incrementamos la cantidad de canicas en el contenedor dependiendo la posición final
        contenedores[posicion] += 1
   
    # Devolver la lista de contenedores con la cantidad de canicas en cada uno
    return contenedores
```

```python
def graficar_histograma(contenedores):
     # Obtenemos el número de niveles y generamos una lista de niveles
    num_niveles = len(contenedores)
    niveles = list(range(num_niveles))

    # Graficamos el histograma utilizando la función bar() de matplotlib
    plt.bar(niveles, contenedores)

     # Agregamos etiquetas a los ejes x e y y un título al gráfico
    plt.xlabel('Contenedores')
    plt.ylabel('Cantidad de canicas')
    plt.title('Histograma de la grafica')

    # Mostramos el gráfico
    plt.show()
```

```python
# Parámetros de la simulación
num_canicas = 3000
num_niveles = 12

# Simulación de la máquina de Galton
resultados = simular_galton(num_canicas, num_niveles)

# Graficamos histograma
graficar_histograma(resultados)
```
