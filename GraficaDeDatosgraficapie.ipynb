```python
import random
import matplotlib.pyplot as plt
```

```python
def simular_galton(num_canicas, num_niveles):
    contenedores = [0] * (num_niveles + 1)

    for _ in range(num_canicas):
        posicion = 0
        for _ in range(num_niveles):
            direccion = random.choice([-1, 1])
            posicion += direccion
        contenedores[posicion] += 1

    return contenedores
```

```python
def graficar_pie(contenedores):
    etiquetas = list(range(len(contenedores)))

    plt.pie(contenedores, labels=etiquetas, autopct='%1.1f%%')
    plt.title('Distribución de canicas en la máquina de Galton')

    plt.show()
```

```python
# Parámetros de la simulación
num_canicas = 3000
num_niveles = 12

# Simulación de la máquina de Galton
resultados = simular_galton(num_canicas, num_niveles)

# Graficar gráfico de tipo "pie"
graficar_pie(resultados)
```
